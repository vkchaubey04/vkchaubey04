**1.** What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot

```python
avg=(df['median_income']).mean()
print('average median income of the data set=',avg)
print()

plt.figure(figsize=(10,7))
plt.hist(df["median_income"],color=['blue'],bins=20)
plt.title("distribution of housing_median_age")
plt.xlabel("--- Range of median_income ---")
plt.ylabel("--- Frequency of median_income ---")
plt.grid(True)
plt.show()

```

* In the dataset most of the values are found between the range of (2,4) of median_income and very less values are present between the range of (12,14)

**2.** Draw an appropriate plot to see the distribution of housing_median_age and explain your observations.

```python
#print(df.describe())
#print()
plt.hist(df["housing_median_age"],color=['blue'],bins=100)
plt.title("distribution of housing_median_age")
plt.xlabel("Range of values")
plt.ylabel("Frequencies")
plt.grid(True)
plt.show()
```

* In a histogram plot, a bin is a range of values into which the data is divided. The data is then grouped into these bins, and the number of data points in each bin is represented by the height of a bar. The bins are typically represented on the x-axis, while the y-axis represents the frequency or count of data points within each bin.
* A small number of bins will result in a coarser representation of the data, while a large number of bins will result in a more detailed representation of the data

```python
plt.boxplot(df["housing_median_age"], showmeans = True,vert=False)
plt.title("Box Plot of housing_median_age")
plt.grid(True)
plt.show()
```

* A box and whisker plot, also known as a box plot, is a standardized way of displaying the distribution of data based on five number summary ("minimum", first quartile (Q1), median, third quartile (Q3), and "maximum"). In Matplotlib, the boxplot() function can be used to create box and whisker plots.
* **showmeans=True** : this parameter is used to display the mean value in box plot.
* **vert=False** : parameter used to rotate box horizontally.

**3.** Show with the help of visualization, how median_income and median_house_values are related?

```python
plt.figure(figsize=(10,7))
plt.scatter(df['median_house_value'],df['median_income'],alpha=0.7,marker='o',edgecolors='black',linewidths=2)
plt.xlabel('median_house_value')
plt.ylabel('median_income')
plt.ylim(0,20)
plt.show()
```

* Scatter plots are useful for visualizing the distribution of the data, as well as any patterns or relationships that may exist between the two variables. 
* Scatter plots can also be used to identify outliers, or data points that are significantly different from the rest of the data. These outliers can be important to identify, as they can indicate a problem with the data or an underlying phenomenon that is not captured by the other data points.

**10.** Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
df['total_bedroom_size']=df['total_bedrooms']*1
df.loc[df['total_bedrooms']<=10,'total_bedroom_size']='SMALL'
df.loc[(df['total_bedrooms']>=11)&(df['total_bedrooms']<1000),'total_bedroom_size']='MEDIUM'
df.loc[df['total_bedrooms']>=1000,'total_bedroom_size']='LARGE'
df.to_csv('new_Dataset.csv',index=True,header=False)
```

* new column is created of "total_bedroom_size" and according to the conditions all rows filled with the value of 'SMALL',"MEDIUM' and 'LARGE' in this new column and also stored in a new dataset i.e **'new_Dataset.csv'**
